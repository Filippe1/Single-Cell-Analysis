```python
# creating list with cells with known doublets: 

donorlist = []

for i in list(rnadata.obs.index): 
    if (i[-1] == '1') & (libone[bonly.obs['cellidmatch'] == i[0:-2]].shape[0] != 0):
        string = libone[bonly.obs['cellidmatch'] == i[0:-2]].iloc[0, 2]
        donorlist.append(string)
        
    elif (i[-1] == '2') & (libtwo[bonly.obs['cellidmatch'] == i[0:-2]].shape[0] != 0):
        string = libtwo[bonly.obs['cellidmatch'] == i[0:-2]].iloc[0, 2]
        donorlist.append(string)
        
    elif (i[-1] == '3') & (libthre[bonly.obs['cellidmatch'] == i[0:-2]].shape[0] != 0):
        string = libthre[bonly.obs['cellidmatch'] == i[0:-2]].iloc[0, 2]
        donorlist.append(string)
        
    elif (i[-1] == '4') & (libfour[bonly.obs['cellidmatch'] == i[0:-2]].shape[0] != 0):
        string = libfour[bonly.obs['cellidmatch'] == i[0:-2]].iloc[0, 2]
        donorlist.append(string)
    
    else:
        donorlist.append('unknown')
        
```
